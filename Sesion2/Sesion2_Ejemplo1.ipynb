```python
# Los datos de tipo dataframe 
# Estos son datos bidimensionales
# 
# Definicion

pd.DataFrame([[2,4,6,8],
              [1,3,5,7],
              [6,6,6,np.nan]])
```

```python
# Etiquetemos las filas (index) y las columnas (columns)
pd.DataFrame([[2,4,6,8],
              [1,3,5,7],
              [6,6,6,np.nan]],
             index = ["f1", "f2", "f3"],
             columns = ["c1", "c2", "c3", "c4"])
```

```python
# Podemos definir un dataframe a partir del etiquetado de cada columna 
pd.DataFrame({"Col1": [10,20,33],
              "Col2": ["Aprob", "Desap", "Aprob"]},
             index = ["Alumno1", "Alumno2", "Alumno3"])
```

```python
# Usemos list comprehensions de python para definir las etiquetas de 
# filas y columnas de algun dataframe 
n_Filas = 102400
n_Columnas = 2056
pd.DataFrame(np.random.randn(n_Filas,n_Columnas),
             index = [f"fila_{_}" for _ in range(1,n_Filas+1)],
             columns = [f"Col_{_}" for _ in range(1,n_Columnas +1)])
```

```python
# Indexing y slicing para objetos bidimensionales de pandas (DataFrames)
# Tenemos 5 formas para realizar estos calculos 
# []
# .loc[]
# .iloc[]
# Indexado a partir de expresiones booleanas
# .query()

df = pd.DataFrame({
    "Nombre" : ["Luis", "Diego", "Madai", "Jorge"],
    "LP" : ["Python", "Julia", "C++", "R"],
    "Nota" : [12,14,16,19]
})

df.dtypes
```

```python
# Indexacion con []
df["Nombre"]
# Devuelve un dato de tipo series
print("El tipo de dato es : " , type(df["Nombre"]))
```

```python
# Construyamos un dataframe seleccionando una sola columna 
df[["Nombre"]]
print("El tipo de dato es : " , type(df[["Nombre"]]))
```

```python
# Seleccion de dos columns 
df[["Nombre", "LP"]]
```

```python
# Error !!!
# Consideremos usar indices enteros
df[0]
```

```python
# Seleccionemos la primera fila
df[0:1]
```

```python
# Deseo todas las filas desde la segunda en adelante
df[1:]
```

```python
# Usemos ahora los metodos .loc y .iloc
# Primera fila 
df.iloc[0]
```

```python
# Recuperemos las primeras dos filas 
df.iloc[0:2]
```

```python
# deseo recuperar el elemento de la cuarta fila / segunda columna
df.iloc[3,1]
```

```python
# Recuperemos la primera y tercera fila con la segunda y tercera columna
df.iloc[[0,2],[1,2]]
```

```python
# El metodo loc sirve cuando puedo usar etiquetas 
# Columnas Nombre
df.loc[:, "Nombre"]
```

```python
# Deseo recuperar dos columnas 
df.loc[:, "LP":"Nota"]
```

```python
# Deseo las dos primeras filas de la columna Nombre
df.loc[[0,1],["Nombre"]]
```

```python
# Indexado a partir de expresiones booleanas 
df[df["Nota"] > 13]
```

```python
# Seleccionamos la informacion qu corresponde a jorge
df[df["Nombre"] == "Jorge"]
```

```python
#  Indexado usando el metodo query
df.query("Nota >=15 & LP == 'Python'")
```
